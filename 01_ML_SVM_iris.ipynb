### 1. module configuration

```python
import pandas as pd
from sklearn.svm import SVC
```

### 2. Data Loader 

```python
csv = pd.read_csv('../data/iris.csv')
csv
```

### 3. Data Split - feature | label

```python
label = csv['variety']
feature = csv[['sepal.length','sepal.width','petal.length','petal.width']]
feature

#label  -->  sklearn는 데이터를 shuffle해줘야 한다.
```

### 4. Model Generate and Training 

```python
model = SVC() # svm 모델을 생성
model.fit(feature,label) # 지도학습이기 때문에 속성과 라벨을 동시에 학습시 넣어준다.
```

### 5. Predict
    학습 때 사용하지 않은 미지의 데이터를 입력해서 예측값을 출력해본다.

```python
pred = model.predict([
    [5.2, 3.4, 1.2, 0.3],
    [2.5, 0.4, 12.3, 4.0],
    [1.2, 6.3, 2.4, 12.8]
    
])

print(pred)
```
