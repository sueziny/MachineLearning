# 1. Load Data
    ScikitLearn(sklearn) DataSet
    유방암 데이터셋
    DecisionTree Classification

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
import numpy as np

cancer = load_breast_cancer()
cancer # 이 안에 유방암 데이터 다 들어있다.
```

# 2. 데이터 구성 및 구조 파악

```python
# 1. Data Feature
X=cancer.data
# X

# 2.Data Label
y=cancer.target
# y

print('Feature Name \n', cancer.feature_names)
print('*'*100)
print('Target Name \n', cancer.target_names)
print('*'*100)
print('Cancer Data \n', cancer.data.shape)
```

```python
dir(cancer)
cancer.keys()
np.bincount(cancer.target)
cancer.target[cancer.target==0].shape

print(cancer.DESCR)
```

# 3. 파이썬의 주요 기능
    - enumerate(), zip(), T(), map()

```python
'''
zip()
파이썬 내장함수
같은 길이를 가진 리스트가 있을 때
같은 인덱스 값끼리 가져와서 사용할 수 있도록 해준다.
'''

name = ['james','Peter','Tomas']
age = [33,44,55]

for n,a in zip(name, age):
    print(n, a)
```

```python
print('1.Sample Count Per class : \n{}'.format(
        {n:v for n, v in zip(cancer.target_names, np.bincount(cancer.target))}
))
```

```python
for i, name in enumerate(cancer.feature_names):
    print('%02d : %s' %(i+1,name))
```

# 4. Train, Test Data Split and Shuffle

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
len(X_train) #455/569
len(X_test) #114/569
```

# 5. Model Generator

```python
dTree = DecisionTreeClassifier()
dTree.fit(X_train,y_train)
dTree
```

```python
pred = dTree.predict(X_test)
pred
```

```python
y_test
```

# 6. Score Evaluation

```python
from sklearn.metrics import accuracy_score

# normalize=True - DataSet에서 True로 정확히 예측한 데이터의 확률
# normalize=False - DataSet에서 True로 정확히 예측한 데이터의 건수
score = accuracy_score(y_test, pred, normalize=True)
print(score)

score2 = accuracy_score(y_test, pred, normalize=False)
print(score2)
```
